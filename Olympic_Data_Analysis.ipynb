```python
sex_counts = df2['Sex'].value_counts()

# Calculate participation rate (assuming total participation is the number of rows in df2)
participation_rate = sex_counts / len(df2) * 100

# Plotting the data
plt.figure(figsize=(8, 6))
sns.barplot(x=participation_rate.index, y=participation_rate.values, palette='Blues_d')

# Adding titles and labels
plt.title('Sex vs Participation Rate', fontsize=16)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Participation Rate (%)', fontsize=14)

# Display the plot
plt.show()
```

```python
city_counts = df2.groupby('City')['Year'].nunique().reset_index(name='Total Years')

# Sort the values for a better visual presentation
city_counts_sorted = city_counts.sort_values(by='Total Years', ascending=False)

print("Total number of years each city hosted the events:")
for index, row in city_counts_sorted.iterrows():
    print(f"{row['City']} hosted the events {row['Total Years']} times.")

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(city_counts_sorted['City'], city_counts_sorted['Total Years'], color='skyblue')

plt.title('Total Number of Years Each City Hosted the Events')
plt.xlabel('Number of Years')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
medals_by_year = df2.groupby('Year')['Medal'].count().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Medal', data=medals_by_year, marker='o')
plt.title('Medals by Year')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.show()
```

```python
medals_by_year = df2.groupby(['Year', 'Medal']).size().unstack(fill_value=0).reset_index()


medals_by_year.columns = ['Year', 'Bronze', 'Gold', 'Silver', 'No Medal']


for index, row in medals_by_year.iterrows():
   print(f"Year: {row['Year']}, Gold: {row['Gold']}, Silver: {row['Silver']}, Bronze: {row['Bronze']}")
```

```python
team_total_medals = df2.groupby('Team')['Medal'].count().reset_index()


total_medals = df2['Medal'].count()


team_total_medals['Winning Rate'] = team_total_medals['Medal'] / total_medals


top_teams_by_win_rate = team_total_medals.sort_values(by='Winning Rate', ascending=False).head(10)


print(top_teams_by_win_rate)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Winning Rate', y='Team', data=top_teams_by_win_rate, palette='viridis')
plt.title('Top 10 Teams by Winning Rate')
plt.xlabel('Winning Rate')
plt.ylabel('Team')
plt.show()
```

```python
# Assuming the user meant to use a DataFrame named df2 with the column 'Team'
team_medal_count = df2.groupby(['Team', 'Medal']).size().unstack(fill_value=0).reset_index()

# Keep only Gold, Silver, and Bronze medals
team_medal_count = team_medal_count[['Team', 'Gold', 'Silver', 'Bronze']]

# Sort by Gold and Silver medals (descending)
top_teams = team_medal_count.sort_values(by=['Gold', 'Silver'], ascending=False).head(10)

print("team_medal_count")
print(team_medal_count)

# Plot
top_teams.plot(x='Team', kind='bar', stacked=True, figsize=(12, 6), color=['#FFD700', '#C0C0C0', '#CD7F32'])  # Gold, Silver, Bronze
plt.title('Top 10 Teams by Medal Count (Gold, Silver, Bronze)')
plt.xlabel('Team')
plt.ylabel('Medal Count')
plt.legend(title='Medal Type')
plt.show()
```

```python
sex_medal_count = df2.groupby(['Sex', 'Medal']).size().unstack(fill_value=0).reset_index()

# Keep only Gold, Silver, and Bronze medals
sex_medal_count = sex_medal_count[['Sex', 'Gold', 'Silver', 'Bronze']]

# Sort by Gold and Silver medals (descending)
top_sex = sex_medal_count.sort_values(by=['Gold', 'Silver'], ascending=False).head(10)

print("sex_medal_count")
print(sex_medal_count)


# Plot
top_sex.plot(x='Sex', kind='bar', stacked=True, figsize=(10, 6), color=['#FFD700', '#C0C0C0', '#CD7F32'])  # Bronze, Gold, Silver colors
plt.title('Medal Count by Gender')
plt.xlabel('Sex')
plt.ylabel('Medal Count')
plt.legend(title='Medal Type')
plt.show()
```

```python
city_medal_count = df2.groupby(['City', 'Medal']).size().unstack(fill_value=0).reset_index()

# Keep only City, Gold, Silver, and Bronze medals
city_medal_count = city_medal_count[['City', 'Gold', 'Silver', 'Bronze']]  # Correct column selection

# Sort by Gold and Silver medals (descending)
top_cities = city_medal_count.sort_values(by=['Gold', 'Silver'], ascending=False).head(10)

print("City Medal Count:")
print(city_medal_count)

# Plot
top_cities.plot(x='City', kind='bar', stacked=True, figsize=(12, 6), color=['#FFD700', '#C0C0C0', '#CD7F32'])  # Gold, Silver, Bronze
plt.title('Top 10 Cities by Medal Count')
plt.xlabel('City')
plt.ylabel('Medal Count')
plt.legend(title='Medal Type')
plt.show()
```
